```python
import torch
import torch.nn.functional as F
```

```python
z=torch.FloatTensor([1,2,3])
y_hat=F.softmax(z, dim=0) #1차원 순서대로 1/6 2/6 3/6
print(y_hat)
print(y_hat.sum())
```

```python
z=torch.rand(3,5, requires_grad=True)
print(z)
y_hat = F.softmax(z, dim=1) #한 행으로 softmax 해줌
print(y_hat)
print(y_hat.sum(dim=1)) #각 행 /각 행을 다 더한 것  
```

```python
y = torch.randint(5, (3,)).long()
print(y) #0부터 4까지의 값을 랜덤하게 3개를 발생시킴
print(y_hat) #첫 번째 값은 1행을 인코딩하고 나머지는 0으로, 두 번째 값은 2행을 인코딩하고 나머지는 0으로...
y_one_hot = torch.zeros_like(y_hat)
print(y_one_hot)
print(y.unsqueeze(1)) #크기가 1인 2차원을 하나 만듦
y_one_hot.scatter_(1, y.unsqueeze(1), 1)
print(y_one_hot)
```
