```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
```

```python
torch.manual_seed(1)
x_data = torch.FloatTensor([[1],[2],[3],[4],[5]]) #입력값
t_data = torch.FloatTensor([[3],[5],[7],[9],[11]]) #정답
```

```python
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear=nn.Linear(1, 1)
        
    def forward(self, x):
        return self.linear(x)
```

```python
model = LinearRegressionModel()
print(model)
optimizer = optim.SGD(model.parameters(), lr=0.01)  #SGD 확률적 그라디언트 디센트
```

```python
epochs = 2001  #epoch 전체 데이터를 한 번 돌리는 것

for epoch in range(epochs):
    y = model(x_data) #이때 forward가 호출됨
    cost = F.mse_loss(y, t_data) #전체에 대한 계산을 한 번에 함
    
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    
    if epoch % 100 == 0: #0번, 100번 ,200번에 한 번씩 찍는 것
        print(epoch, cost.item(), list(model.parameters()))
        
```

```python
predict = (model(torch.FloatTensor([[7], [8]]))) #데이터는 2차원 형태로 들어가야 함 #이때 forward가 호출됨
print(predict) 
```

```python
import pandas as pd
df=pd.read_csv('../data/data-01-test-score.csv', header=None) #header=None을 해야 첫 번째 데이터가 헤더로 안 들어감
x_data = df.values[:, :-1] #values를 넣으면 배열 형태로 출력됨 , 0부터 마지막열 앞까지
t_data = df.values[:, -1]
print(x_data.shape, t_data.shape)
t_data=t_data.reshape(-1, 1)
print(x_data.shape, t_data.shape)
print(type(x_data), type(t_data)) #훈련시키려면 텐서플로우로 만들어 주어야 함
x_data1 = torch.from_numpy(x_data).float()
t_data1 = torch.from_numpy(t_data).float()
print(type(x_data1), type(t_data1)) #훈련데이터를 출력함

indim = x_data1.size(1)
print(indim)
```

```python
class MultiLinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear= nn.Linear(indim,1) #size1의 크기를 변수로 받아서, 그 변수를 숫자 대신에 입력해도 됨
    def forward(self, x):
        return self.linear(x)
```

```python
model = MultiLinearRegressionModel()
optimizer = optim.SGD(model.parameters(), lr = 1e-6)
```

```python
epochs = 200001
for epoch in range(epochs):
    y = model(x_data1)
    cost = F.mse_loss(y, t_data1)
    
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    
    if epoch % 100 == 0:
        print(epoch, cost.item(), list(model.parameters()))
```
